```python
database["Known_For22"].value_counts()
```

Since quite a few restaurants don't have anything that they are known for, I'll fill them with "no reviews" values


```python
database["Known_For22"].fillna("No reviews", inplace=True)
```

```python
database.head()
```

Our last column to be cleaned is the Delivery_Rating
Simply putting 0 for the null values would lead us to an ill-represented data, since I am sure the restaurants would have at least some rating.

So what I'll do instead, is to give them a 'mean' value instead. That sounds like the most appropriate way to fill the columns.


```python
database["Delivery_Rating"].fillna(database["Delivery_Rating"].mean(), inplace=True)
```

```python
database.head()
```

Q2.Which restaurant is the best one having ratings more than 4.5 for delivery


```python
database[database.Delivery_Rating>4.5]
```

```python
database[database.Delivery_Rating==4.5]
```

There aren't any datasets with delivery rating above 4.5
so we shall operate on delivery ratings of 4.5 itself

The columns useful for comparison are pricing for 2 and delivery rating count. It gives an estimate of how affordable and popular the restaurant is.

```python
data=database[database.Delivery_Rating==4.5]
data.reset_index(inplace=True)
data.head(10)
```

```python
x=num.arange(8)
y2=data["Pricing_for_2"]
y3=data["Delivery_Rating_Count"]
width=0.2
plt.bar(x-0.2,y2,width,color="r")
plt.bar(x+0.2,y3,width,color="g")

plt.xlabel("Restaurants")
plt.ylabel("Number")

plt.xticks(x,data["Restaurant_Name"],rotation='vertical')
plt.legend(["Pricing_for_2","Delivery_Rating_Count"])
plt.show()
```

as we can see, the lowest prices and the most reviews have been received by natural ice cream, and thus we can deduce it to be the best restaurant for deliveries

Q5.Which restaurant is the best one having ratings more than 4.5 for dining?

```python
data_2=database[database["Dining_Rating"]>4.5]
data_2
```

Now, we shall compare based off the pricing for 2 vs number of reviews received.

```python
data_2=data_2[data_2['Pricing_for_2']<data_2['Pricing_for_2'].mean()]
data_2=data_2[data_2['Dining_Review_Count']>data_2['Dining_Review_Count'].mean()]
data_2
```

Now that we have found out the restaurants who have more reviews than the mean value of reviews per restaurant AND in such restaurants, those only which have prices less than the mean value of prices per restaurant, we can finally deduce the best restaurant easily.

```python
data_2.reset_index(inplace=True)
data_2.head(n=22)
```

```python
z=num.arange(21)
y2=data_2["Pricing_for_2"]
y3=data_2["Dining_Review_Count"]
width=0.2
plt.bar(z-0.2,y2,width,color="r")
plt.bar(z+0.2,y3,width,color="g")

plt.xlabel("Restaurants")
plt.ylabel("Number")

plt.xticks(z,data_2["Restaurant_Name"],rotation='vertical')
plt.legend(["Pricing_for_2","Dining_Review_Count"])
plt.show()
```

AND BAM! There it is.
Saravana Bhavan with greatest number of reviews and lowest prices. Our winner
